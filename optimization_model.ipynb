```python
# IQR 기준 이상치 확인하는 함수
def findOutliers(x,column):
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    iqr = q3 - q1
    y = x[(x[column] > (q3 + 3*iqr)) | (x[column] < (q1 - 3*iqr))]
    return len(y)
# price, duration, days_left에 대하여 IQR 기준 이상치 개수 확인

for i in range(1, train_df.shape[1]):
    col = train_df.columns[i]
    print(f"{col} iqr outlier", findOutliers(train_df, col))
```

```python
# IQR 기준 이상치 확인하는 함수
def changeOutliers(x,column):
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)

    iqr = q3 - q1

    # 이상치를 대체할 Min, Max 설정
    Min = q1 - 3*iqr
    Max = q3 + 3*iqr

    # Max 보다 큰 값을 Max로 , Min보다 작은 값을 Min으로 대체
    x.loc[(x[column] > Max), column] = Max
    x.loc[(x[column] < Min), column] = Min

    return(x)

for i in range(1, train_drop.shape[1]):
    col = train_drop.columns[i]
    clean_train_df = changeOutliers(train_drop, col)

for i in range(1, clean_train_df.shape[1]):
    col = clean_train_df.columns[i]
    print(f"{col} iqr outlier", findOutliers(clean_train_df, col))
```

```python
#VIF 구하기

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor


# VIF 출력을 위한 데이터 프레임 형성
vif = pd.DataFrame()

# VIF 값과 각 Feature 이름에 대해 설정
vif["VIF Factor"] = [variance_inflation_factor(clean_train_df.values, i) for i in range(clean_train_df.shape[1])]
vif["features"] = clean_train_df.columns

# VIF 값이 높은 순으로 정렬
vif = vif.sort_values(by="VIF Factor", ascending=False)
vif = vif.reset_index().drop(columns='index')
vif
```

```python
from sklearn.model_selection import train_test_split

x = train_pca_y.drop(columns=['y'], axis=1)
y = train_pca_y['y']
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=1)
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
```

```python
param_xgb = {"max_depth": [10,30,50],
              "min_child_weight" : [1,3,6,10],
              "n_estimators": [100,200,300,500,1000]
              }
xgb = XGBRegressor(random_state=1)
grid_search = GridSearchCV(estimator=xgb, param_grid=param_xgb, cv=5)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
y_pred = grid_search.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
np.sqrt(MSE)
```
